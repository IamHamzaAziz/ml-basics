# Using Linear Regresion find Per Capita Income in Canda in 2020

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Gicing new names to columns and skipping second row because it had previous column names
df = pd.read_csv('canada_per_capita_income.csv', names=['year', 'income'], skiprows=1)
df.head(3)
```

```python
reg = linear_model.LinearRegression()

# Giving df.year (in 2D array format) and df.income to reg so that it can understand the data
# And give us future predictions
reg.fit(df[['year']], df.income)
```

```python
# Since year was independent variable and in x axis while income was dependent and in y variable
# So here we passed a year value 2020 and asked to give income prediction
reg.predict([[2020]])
```

```python
plt.xlabel('Year')
plt.ylabel('Income USD $')
plt.title('Per Capita Income in Canda')
plt.scatter(df.year, df.income)

# Slop line showing Linear Regression line
# In x axis we have df.year and in y axis we ask prediction of each year as per data
# Or prediction made by formula y=m*x+b where m is slope and b is y intercept
plt.plot(df.year, reg.predict(df[['year']]), color='r')
```
